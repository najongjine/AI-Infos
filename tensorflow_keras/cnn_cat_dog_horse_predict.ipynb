```python
# 1️⃣ Colab 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# 2️⃣ 필요한 모듈 임포트
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# 3️⃣ 저장된 모델 불러오기
model_path = '/content/drive/MyDrive/my_models/tensorflow_keras/cnn_vgg16_cat_dog_horse.h5'
model = load_model(model_path)
print("✅ 모델 불러오기 성공!")

# 4️⃣ 사용자로부터 사진 입력받기
from google.colab import files
uploaded = files.upload()  # 사용자 업로드창 뜸

# 업로드된 파일 이름 가져오기
img_filename = list(uploaded.keys())[0]
print(f"✅ 입력된 이미지: {img_filename}")

# 5️⃣ 이미지 불러와서 전처리
IMG_HEIGHT = 150
IMG_WIDTH = 150

img = image.load_img(img_filename, target_size=(IMG_HEIGHT, IMG_WIDTH))
img_array = image.img_to_array(img)
print(f"img_array type:{type(img_array)}") # <class 'numpy.ndarray'>
img_array = img_array / 255.0  # 정규화
img_array = np.expand_dims(img_array, axis=0)  # 배치 차원 추가 (1장짜리)

# 6️⃣ 예측하기
predictions = model.predict(img_array)

# 예측 결과 출력
class_names = ['cat', 'dog', 'horse']  # 너가 학습할 때 사용한 클래스 순서
predicted_class = class_names[np.argmax(predictions)]

print(f"✅ 예측 결과: {predicted_class}")
print(f"✅ 클래스별 확률: {predictions}")

# 7️⃣ 이미지도 함께 보여주기
plt.imshow(img)
plt.axis('off')
plt.title(f"Prediction: {predicted_class}")
plt.show()
```
